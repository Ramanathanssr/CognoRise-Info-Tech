```python
avg_unemployment_rate=df.groupby('Region')['Estimated Unemployment Rate (%)'].mean()
state_with_highest_unemployment_rate=avg_unemployment_rate.idxmax()
highest_unemployment_rate=avg_unemployment_rate.max()
state_with_lowest_unemployment_rate=avg_unemployment_rate.idxmin()
lowest_unemployment_rate=avg_unemployment_rate.min()
```

```python
print('state_with_highest_unemployment_rate:')
print(state_with_highest_unemployment_rate,':',highest_unemployment_rate,'\n')
print('state_with_lowest_unemployment_rate:')
print(state_with_lowest_unemployment_rate,':',lowest_unemployment_rate)
```

```python
print('Average Unemployment Rate:',avg_unemployment_rate)
```

```python
plt.figure(figsize=(10,8))
plt.barh(avg_unemployment_rate.index,avg_unemployment_rate.values,color='green')
plt.title('Average Unemployment Rate by Region')
plt.xlabel('Average unemployment rate')
plt.ylabel('Region')
plt.show()
```

# Exploratory Data Analysis

```python
# Value counts of work year:
year_count=df2['work_year'].value_counts()

# EDA for work Year:
plt.figure(figsize=(8,4))
plt.bar(year_count.index,year_count.values,color=sns.color_palette(),edgecolor='black')
plt.title('Value counts of work year')
plt.xlabel('Year')
plt.ylabel('Count')
plt.show()
```

```python
# Value counts of Experience Level:
experience_count=df2['experience_level'].value_counts()

# EDA for Experience Level:
plt.figure(figsize=(10,4))
plt.bar(experience_count.index,experience_count.values,color=sns.color_palette(),edgecolor='black')
plt.title('Distribution of Experience level')
plt.xlabel('Experience_Level')
plt.ylabel('Count')
plt.show()
```

```python
job_title_count=df2['job_title'].value_counts()
job_title_count
```

```python
# value counts of employment types:
employment_count=df2['employment_type'].value_counts()

# EDA for Employment types:
plt.figure(figsize=(10,5))
plt.bar(employment_count.index,employment_count.values,color=sns.color_palette(),edgecolor='black')
plt.title('Distribution of Employment Types')
plt.xlabel('employment_types')
plt.ylabel('counts')
plt.show()
```

```python
# Value counts of salary currency:
salary_currency_count=df2['salary_currency'].value_counts()

# EDA for salary currency count:
plt.figure(figsize=(10,5))
plt.bar(salary_currency_count.index,salary_currency_count.values,color=sns.color_palette(),edgecolor='black')
plt.title('Distribution of Salary Currency')
plt.xlabel('Salary Currency')
plt.ylabel('Count')
plt.show()
```

```python
# Value counts of Remote Ratio:
remote_ratio_counts=df2['remote_ratio'].value_counts()

# EDA for Remote Ratio:
plt.figure(figsize=(8,4))
plt.bar(remote_ratio_counts.index,remote_ratio_counts.values,color=sns.color_palette(),edgecolor='black',width=25)
plt.title('Distribution of Remote Ratio')
plt.xlabel('Remote Ratio')
plt.ylabel('count')
plt.show()
```

```python
# Value Counts of Company size:
company_size_count=df2['company_size'].value_counts()

# EDA for Company Size:
plt.figure(figsize=(8,4))
plt.bar(company_size_count.index,company_size_count.values,color=sns.color_palette(),edgecolor='black')
plt.title('Distribution of Company Size')
plt.xlabel('Company Size')
plt.ylabel('Count')
plt.show()
```

```python
job_wise_salary=df2.groupby('job_title')['salary'].sum()

# EDA for Job title and their salary:
plt.figure(figsize=(10,5))
plt.bar(job_wise_salary.index,job_wise_salary.values,color=sns.color_palette(),edgecolor='black')
plt.title('Job Title & Their Salary')
plt.xlabel('Job Title')
plt.ylabel('Salary')
plt.xticks(rotation=90)
plt.show()
```

```python
experience_level_salary=df2.groupby('experience_level')['salary'].mean()
print(' EN-Entry Level\n MI-Mid Level\n SE-Senior Level\n EX-Executive Level')

# EDA for Experience Level salary:
plt.figure(figsize=(10,5))
plt.pie(x=experience_level_salary,labels=experience_level_salary.index,colors=sns.color_palette(),autopct='%1.1f%%',startangle=90,explode=[0.05]*len(experience_level_salary))
plt.title('Distribution of Experience Level & Their Salary')
plt.show()
```

```python
company_size_salary=df2.groupby('company_size')['salary'].mean()
print(' S-Small\n M-Medium\n L-Large')

# Plot for Company Size and their Salary:
plt.figure(figsize=(10,5))
plt.pie(x=company_size_salary,labels=company_size_salary.index,colors=sns.color_palette(),autopct='%1.1f%%',startangle=180,explode=[0.05]*len(company_size_salary))
plt.title('Company Size and Their Salary')
plt.show()
```
